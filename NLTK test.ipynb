```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import pandas as pd
import requests
from IPython.display import display, HTML

display(HTML("<h3><i><u>Test data pull from imdb csv to create dataframe</u></i></h3>"))
```

```python
def callimdblist(): #pull data from csv into dataframe
    imdb_top_250_english = "IMDB_Top250Engmovies2_OMDB_Detailed.csv"
    data = pd.read_csv(imdb_top_250_english)
    return data

if __name__ == "__main__": #unneccessary - just wanted to test it
    data = callimdblist()
    data = data[['Rank', 'Title', 'Year', 'Rated', 'Genre', 'Director', 'Actors', 'Plot']]
    #refine dataset
    data['Tokens'] = ''
    for index, row in data.iterrows(): #for each plot, tokenize and remove stopwords
        movie_plot = row['Plot']
        tokenized_list = word_tokenize(movie_plot)
        list_without_punctuation = [word for word in tokenized_list if word.isalpha]
        stop_words = set(stopwords.words('english'))
        list_without_stopwords = [word for word in list_without_punctuation if not word in stop_words]
        data.at[index, 'Tokens'] = list_without_stopwords
    display(data.head())
    
    
text = input("Enter a key word to search for: ")
print("\n")
for index, row in data.iterrows(): #print titles of the films that 
                                    #match the search term in the plot
    if text in row['Tokens']:
        display(f"{row['Title']}")
```
