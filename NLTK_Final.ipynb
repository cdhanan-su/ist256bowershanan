```python
def callimdblist(): #pull data from csv into dataframe
    imdb_top_250_english = "IMDB_Top250Engmovies2_OMDB_Detailed.csv"
    data = pd.read_csv(imdb_top_250_english)
    return data



def create_table_250():
    data = callimdblist()
    data = data[['Rank', 'Title', 'Year', 'Rated', 'Genre', 'Director', 'Actors', 'Plot']]
    #refine dataset
    data['Tokens'] = ''
    for index, row in data.iterrows(): #for each plot, tokenize and remove stopwords
        movie_plot = row['Plot']
        tokenized_list = word_tokenize(movie_plot)
        list_without_punctuation = [word for word in tokenized_list if word.isalpha]
        stop_words = set(stopwords.words('english'))
        list_without_stopwords = [word for word in list_without_punctuation if not word in stop_words]
        data.at[index, 'Tokens'] = list_without_stopwords
    return data
    
    
    
def search_params(data):
    while True:
        choice = input("Would you like to search by: plot, actor, director, or genre? ")
        choice = str.lower(choice)
        if choice == "plot":
            column = "Tokens"
            term = input("Please enter a single word to search for: ")
            search_func(term, column, data)
            break
        elif choice == "actor":
            column = "Actors"
            term = input("Please enter a single name to search for: ")
            search_func(term, column, data)
            break
        elif choice == "director":
            column = "Director"
            term = input("Please enter a single name to search for: ")
            search_func(term, column, data)
            break
        elif choice == "genre":
            column = "Genre"
            term = input("Please enter a single word to search for: ")
            term = str.capitalize(term)
            search_func(term, column, data)
            break
        else:
            print("\nYou did not make a valid selection. Please enter 'plot', 'actor', 'director', or 'genre'. \n")
    menu_route = continuation_func(data)
    return menu_route   

            
            
def search_func(search, column, table): 
    print("\nSearch Results: ")
    for index, row in table.iterrows(): #print titles of the films that 
                                    #match the search term in the plot
        if search in row[column]:
            display(f"{row['Title']}")
    return
                    

def continuation_func(data):
    while True:
        print("Would you like to: \n\n    Learn more about one of these titles? Enter '1' to explore. ")
        print("    Search the top 250 again? Enter '2' to search again. ")
        print("    Return to the main menu? Enter '3' to continue. ")
        print("    Quit. Enter 'q' to quit. \n")
        choice = input()
        if choice == '1':
            menu_route = False
            print("Connection to TasteDive_Lookup")
            break
        elif choice == '2':
            menu_route = False
            search_params(data)
            break
        elif choice == '3':
            menu_route = True
            break
        elif choice == 'q':
            menu_route = False
            break
        else:
            print("Please make a valid selection. Choose '1' to explore, '2' to search again, '3' to return, or 'q' to quit. ")
        
    return menu_route
```
